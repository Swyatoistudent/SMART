## ARIMA

```python
from pmdarima.arima import auto_arima
```

```python
all_cat_df = all_cat_df.resample('D').ffill()
train = all_cat_df[:650]
valid = all_cat_df[650:]
```

```python
valid
```

```python
all_cat_df.info()
```

```python
model = auto_arima(train,suppress_warnings=True)
model.fit(train)
forecast = model.predict(n_periods=len(valid))
forecast = pd.DataFrame(forecast,index = valid.index,columns=['Prediction'])
```

## ML model(XGboost)

```python
def create_features(df,lab):
    df['day_of_week']=df['date'].dt.day_of_week
    df['quarter']=df.date.dt.quarter
    df['month']=df.date.dt.month
    df['year'] = df.date.dt.year
    df['dayofyear'] = df.date.dt.dayofyear
    df['dayofmonth'] = df.date.dt.day
    
    X = df[['day_of_week','quarter','month','year',
           'dayofyear','dayofmonth']]
    
    y = df[lab]
    return X, y
```

```python
all_cat_df  = df.groupby(['date']).date.count().reset_index(name='sales')
```

```python
train = all_cat_df[20:550]
valid = all_cat_df[550:]
```

```python
X_train,y_train = create_features(train,'sales')
X_valid,y_valid = create_features(valid,'sales')
```

```python
y_train.info()
```

```python
from xgboost import XGBRegressor
```

```python
XGB_allcat = XGBRegressor(n_estimators = 100)
XGB_allcat.fit(X_train,y_train,eval_metric = 'mae')
```

```python
y_predict = XGB_allcat.predict(X_valid)
```

```python

fig,ax = plt.subplots(figsize=(10,5))
ax.plot(train['date'],train['sales'], c='#004a8f')
ax.plot(valid['date'],valid['sales'], c='#91aac9',linewidth=0.8)
ax.plot(valid['date'],y_predict,c='red',linewidth=0.8)
ax.legend(["train",'valid','xgboost'])
plt.show()
```

```python

fig,ax = plt.subplots(figsize=(10,5))
ax.plot(valid['date'],valid['sales'], c='#91aac9',linewidth=0.8)
ax.plot(valid['date'],y_predict,c='red',linewidth=0.8)
ax.legend(['valid','xgboost'])
plt.show()
```
